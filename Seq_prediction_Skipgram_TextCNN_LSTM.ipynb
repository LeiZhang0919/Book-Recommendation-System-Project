```python
from nltk import ngrams
def get_samples(seq, window_size): # 比这个短的直接没了
    n_grams_samples = []
    stopmark = len(seq) - window_size
    for idx, book_gram in enumerate(ngrams(seq, window_size)):
        if idx == stopmark:
            break
        X = [str(i) for i in book_gram]
        y = str(seq[idx+len(book_gram)])
        n_grams_samples.append([X, y])
    return n_grams_samples
```

```python
ngram_samples = []
ignore_set=0

for seq_list in tqdm(df_user_books_seq['book_id_seq'].apply(lambda x: eval(x)).tolist()):
    tmp_sample = get_samples(seq_list, window_size)
    if len(tmp_sample)>=1: 
        ngram_samples.extend(tmp_sample)
    else:
        ignore_set+=1 # how many user information drop 
```

```python
import unicodedata
import re
from collections import Counter

def unicode_to_ascii(s):
    return ''.join(
        c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) != 'Mn')

def normalize_string(s):
    s = unicode_to_ascii(s.lower())
    s = re.sub(r'http\S+', r' ', s)
    # s = re.sub(r'([!.?])', r' \1', s)
    # s = re.sub(r'([!.?])', r' \1', s)
    s = re.sub(r'[^a-zA-Z]+', r' ', s)
    s = re.sub(r'\s+', r' ', s)
    return s
```
